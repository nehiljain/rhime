```python
import json
from pprint import pprint
import pandas as pd
```

```python
json_resp = None
with open("pocket_data_response.json") as f:
    json_resp = json.load(f)
```

```python
pprint(json_resp['list'])
```

```python
l = []
for k,v in json_resp['list'].items():
    pprint(v)
    l.append(v)
```

```python
len(l)
```

```python
df = pd.DataFrame.from_records(l, columns=["authors","word_count", "time_updated", "time_read", "time_favorited", 
                                           "time_added","sort_id","resolved_title", "resolved_url",
                                          "given_url","given_tile","favorite", "status"])
```
